```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Step 2: Load dataset
df = pd.read_csv("online_retail_II.csv", encoding='ISO-8859-1')

# Step 3: Basic cleaning
df.dropna(subset=['Invoice', 'StockCode', 'Description', 'Quantity', 'Price', 'Customer ID'], inplace=True)
df = df[df['Quantity'] > 0]
df = df[df['Price'] > 0]
df['Invoice'] = df['Invoice'].astype(str)

# Step 4: Simulate sections using Description keywords
def map_section(desc):
    desc = desc.lower()
    if "light" in desc: return "Lighting"
    elif "lamp" in desc: return "Lighting"
    elif "candle" in desc: return "Decor"
    elif "sofa" in desc or "curtain" in desc: return "Furniture"
    elif "mug" in desc or "cup" in desc: return "Kitchenware"
    elif "bag" in desc: return "Fashion"
    else: return "Misc"

df['Section'] = df['Description'].apply(map_section)

# Step 5: Create sequence per invoice
invoice_sections = df.groupby('Invoice')['Section'].apply(list)

# Step 6: Create features: Use first (n-1) items to predict the nth section
sequences = []
for session in invoice_sections:
    if len(session) > 1:
        for i in range(len(session)-1):
            sequences.append((session[:i+1], session[i+1]))

# Turn into dataframe
seq_df = pd.DataFrame(sequences, columns=['PreviousSections', 'NextSection'])

# Convert list to single string as feature
seq_df['Feature'] = seq_df['PreviousSections'].apply(lambda x: ' > '.join(x))

# Step 7: Encode features
X_raw = seq_df['Feature']
y = seq_df['NextSection']

# Encode text using LabelEncoder (simplest way)
feature_encoder = LabelEncoder()
target_encoder = LabelEncoder()
X_encoded = feature_encoder.fit_transform(X_raw)
y_encoded = target_encoder.fit_transform(y)

# Step 8: Split dataset
X_train, X_test, y_train, y_test = train_test_split(X_encoded.reshape(-1, 1), y_encoded, test_size=0.3, random_state=42)

# Step 9: Train multiple models
models = {
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Logistic Regression": LogisticRegression(max_iter=1000)
}

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = classification_report(y_test, y_pred, output_dict=True, zero_division=0)['weighted avg']['f1-score']
    results[name] = {"Accuracy": acc, "F1-score": f1}
    print(f"\n{name} Classification Report:\n")
    print(classification_report(y_test, y_pred, target_names=target_encoder.classes_))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

# Step 10: Plot model comparison
result_df = pd.DataFrame(results).T
result_df.plot(kind='bar', figsize=(10, 5), title="Model Performance Comparison")
plt.ylabel("Score")
plt.ylim(0, 1)
plt.grid(True)
plt.show()
```
